1. Given a stack of integers, sort it in ascending order using another temporary stack. (Hanya boleh menggunakan dua stack)

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class Stack:
  def __init__(self):
    self.top = None

  def push(self, data):
    new_node = Node(data)
    if self.top == None:
      self.top = new_node
    else:
      new_node.next = self.top
      self.top = new_node

  def pop(self):
    if self.top == None:
      return None
    else:
      popped = self.top.data
      self.top = self.top.next
      return popped

  def pushsort(self, data):
    stack2 = Stack()
    new_node = Node(data)
    if self.top == None and stack2.top == None:
      self.top = new_node
    else:
      if self.top.data > new_node.data:
        new_node.next = self.top
        self.top = new_node
      else:
        while self.top != None and self.top.data < new_node.data:
          stack2.push(self.top.data)
          self.pop()
        new_node.next = self.top
        self.top = new_node
        while stack2.top != None:
          self.push(stack2.top.data)
          stack2.pop()

  def print_stack(self):
    temp = self.top
    while temp:
      if temp.next != None:
        print(temp.data, end=", ")
        temp = temp.next
      else:
        print(temp.data, end="")
        break
    print()


stack1 = Stack()
elements = input("Input List: ").split()
# ADD TO STACK
for e in elements:
  stack1.pushsort(int(e))
print("Output: ")
stack1.print_stack()
```
