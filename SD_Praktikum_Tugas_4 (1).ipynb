1. Given an unsorted doubly linked list containing n nodes. The problem is to remove
duplicate nodes from the given list.
image.png

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
    self.prev = None

class LinearDLL:
  def __init__(self):
    self.head = None
    self.tail = None

  # ADD
  def add(self, data):
    new_node = Node(data)
    if self.head == None:
      self.head = new_node
      self.tail = new_node
    else:
      temp = self.head
      while temp.next:
        temp = temp.next
      temp.next = new_node
      new_node.prev = temp
      self.tail = new_node
      new_node.next = None


  # DELETE DUPLICATE
  def delduplicate(self):
      start = self.head
      while start != None and start.next != None:
        temp = start
        while temp.next != None:
          if start.data == temp.next.data:
            duplicate = temp.next
            after = duplicate.next
            temp.next = after
            if after != None:
              after.prev = temp
          else:
              temp = temp.next
        start = start.next


  def print(self):
      temp = self.head
      while temp:
        if temp.next != None:
          print(temp.data, end="<->")
          temp = temp.next
        else:
          print(temp.data, end="")
          break


DLL = LinearDLL()
# DLL.add(8)
# DLL.add(4)
# DLL.add(4)
# DLL.add(6)
# DLL.add(4)
# DLL.add(8)
# DLL.add(4)
# DLL.add(10)
# DLL.add(12)
# DLL.add(12)
elements = input("Input List: ").split()
for e in elements:
  DLL.add(e)
print("Input: ")
DLL.print()
print()
print("Output: ")
DLL.delduplicate()
DLL.print()
```

2. Given a Doubly Linked List, the task is to reverse the given Doubly Linked List.

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
    self.prev = None

class LinearDLL:
  def __init__(self):
    self.head = None
    self.tail = None

  # ADD
  def add(self, data):
    new_node = Node(data)
    if self.head == None:
      self.head = new_node
      self.tail = new_node
    else:
      temp = self.head
      while temp.next:
        temp = temp.next
      temp.next = new_node
      new_node.prev = temp
      self.tail = new_node
      new_node.next = None

  # REVERSED
  def reversed(self):
      if self.head is None or self.head.next == self.head or self.head.prev == self.head:
        return
      before = None
      temp = self.head
      while (temp != None):
        after = temp.next
        temp.next = before
        temp.prev = after
        before = temp
        temp = after
        if after == None:
          break
      self.head = before


  def print(self):
      temp = self.head
      while temp:
        if temp.next != None:
          print(temp.data, end="<->")
          temp = temp.next
        else:
          print(temp.data, end="")
          break

DLL = LinearDLL()
# DLL.add(1)
# DLL.add(2)
# DLL.add(3)
elements = input("Input List: ").split()
for e in elements:
  DLL.add(e)
print("Input: ")
DLL.print()
print()
print("Output: ")
DLL.reversed()
DLL.print()
```
