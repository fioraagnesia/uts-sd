Buatlah sebuah program menggunakan stack dan queue untuk memecahkan masalah berikut!

Pertama diinputkan n, kemudian inputkan bilangan sebanyak n kali, aturan dari program:
*   bilangan yang diinputkan harus dimasukkan ke dalam sebuah queue
*   setelah selesai disimpan ke dalam queue, keluarkan bilangan tersebut dan masukkan ke dalam stack satu per satu dengan aturan sbg berikut:
  *   jika stack kosong, maka bilangan langsung dimasukkan ke dalam stack
  *   jika bilangan yang dikeluarkan dari queue ditambah dengan bilangan teratas stack merupakan bilangan prima, maka bilangan yang dikeluarkan dari queue tidak dimasukkan ke dalam stack, dan bilangan yang teratas dari stack dikeluarkan dari stack
  *   jika bilangan yang dikeluarkan dari queue ditambah dengan bilangan teratas dari stack hasilnya tidak prima, maka masukkan bilangan tersebut ke stack
  *   jika pada akhir program stack kosong atau berisi 1 bilangan, maka outputkan "KOSONG", jika kondisi tidak terpenuhi, maka outputkan "TIDAK KOSONG"
  *   buatlah class stack dan queuenya sendiri (tidak menggunakan library bawaan python)

Contoh:
*   Input: 4
*   Input Bil. 1: 3
*   Input Bil. 2: 2
*   Input Bil. 3: 6
*   Input Bil. 4: 5

Output:
*   Q: 3 2 6 5
*   S:
--------------
*   Q: 2 6 5
*   S: 3
--------------
*   Q: 6 5
*   S:
--------------
*   Q: 5
*   S: 6
--------------
*   Q:
*   S:
--------------
"KOSONG"

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
    self.prev = None


class Queue:
  def __init__(self):
    self.front = None
    self.back = None
    self.counter = 0

  def enqueue(self, data):
    new_node = Node(data)
    if self.front == None:
      self.front = new_node
    else:
      self.back.next = new_node
    self.back = new_node
    self.counter += 1

  def dequeue(self):
    if self.front == None:
      return None
    temp = self.front
    self.front = self.front.next
    if self.front is None:
      self.back = None
    self.counter -= 1
    return temp.data

  def print_queue(self):
    temp = self.front
    while temp:
        print(temp.data, end=" ")
        temp = temp.next
    print()


class Stack:
  def __init__(self):
    self.top = None

  def push(self, data):
    new_node = Node(data)
    if self.top == None:
      self.top = new_node
    else:
      new_node.next = self.top
      self.top = new_node

  def pop(self):
    if self.top == None:
      return None
    else:
      popped = self.top.data
      self.top = self.top.next
      return popped

  def print_stack(self):
    temp = self.top
    while temp:
        print(temp.data, end=" ")
        temp = temp.next
    print()


class QueueStack:
  # Untuk cek bilangan prima
  def cekprima(self, x):
    if x < 2:
      return False
    for i in range(2, x):
      if x % i == 0:
        return False
    return True

  # PROGRAM YANG DIMINTA
  def program(self, elements):
    queue = Queue()
    stack = Stack()

    # Memasukkan input ke queue
    for e in elements:
      queue.enqueue(e)

    # Print Q dan S di awal
    print("Q: ", end="")
    queue.print_queue()
    print("S: ", end="")
    stack.print_stack()
    print()

    # Looping selama queue ada isinya
    while queue.counter > 0:
      temp = queue.dequeue()

      # Jika stack masih kosong, bilangan langsung dimasukkan ke stack
      if stack.top == None:
        stack.push(temp)
      else:
        # Cek hasil penjumlahan prima atau tidak
        topstack = stack.top.data
        x = topstack + temp
        if self.cekprima(x):
          stack.pop()
        else:
          stack.push(temp)

      # Print hasil setelah dipindah
      print("Q: ", end="")
      queue.print_queue()
      print("S: ", end="")
      stack.print_stack()
      print()

    # Cek kondisi terakhir setelah queue habis
    if stack.top == None or stack.top.next == None:
      print("KOSONG")
    else:
      print("TIDAK KOSONG")

qs = QueueStack()
n = int(input("Input: "))
elements = []
for i in range(n):
  print("Input bil.", i+1, ": ", end=" ")
  elements.append(int(input()))
print()
print("Output:")
qs.program(elements)
```
